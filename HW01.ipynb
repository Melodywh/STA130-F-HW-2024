7. What's the difference between "df.dropna()" and "del df['col']"?

- "del df['col']" is used to delete a specific colume which you have to choose by yourself, yet "df.dropna()" automatically picks out the columes that contain missing data for you.

Q7-1. Provide an example of a "use case" in which using df.dropna() might be peferred over using del df['col']

- Ans: 
When a researcher is trying to collect datas from a google form, he/she might want to get rid of the insufficient responses (for example, some respondents may have skipped questions, which will be resulting in "Nan"). However, he/she wouldn't want to delet the whole colume of the answer. In this case, df.dropna() is more preferred to be used.

```python
import pandas as pd
import numpy as np

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'ID': [101, 102, 103, 104],
        'Age': [25, 30, 35, 40],
        'Q1': [5, np.nan, 7, 8],   # Question 1 responses
        'Q2': [5, 6, 7, 8],  # Question 2 responses
        'Q3': [8, 9, np.nan, 10]}  # Question 3 responses

df = pd.DataFrame(data)
print(df)
```

```python
# remove the 'ID' column because it is irrelevant
clean_df = df.drop('ID', axis=1)
print(clean_df)
```

```python
#remove the columes that contain missing data 
clean_df = df.dropna()
print(clean_df)
```
